```python
#In this assignment students have to transform iris data into 3 dimensions
#and plot a 3d chart with transformed dimensions and colour each data
#point with specific class.
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
import seaborn as sns

iris_data=sns.load_dataset('iris')
iris_data
#new_iris=iris_data.drop(['species'],axis=1)
#new_iris=new_iris.drop(['petal_width'],axis=1)

#new_iris
iris_data.head()
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

x = iris_data['sepal_length']
y = iris_data['sepal_width']
z = iris_data['petal_length']
ax.scatter(x, y, z, s=100, alpha=.6, edgecolors='black',color='red',marker='o')

ax.set_xlabel('Sepal Length')
ax.set_ylabel('Sepal Width')
ax.set_zlabel('Petal Length')

plt.show()
```
