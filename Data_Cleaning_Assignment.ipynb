```python
import numpy as np
import pandas as pd
df = pd.DataFrame(
{'From_To': ['LoNDon_paris', 'MAdrid_miLAN','londON_StockhOlm','Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )','12. Air France', '"Swiss Air"']
}
)
df
```

```python
#1. Some values in the the FlightNumber column are missing. These numbers are meant to increase by 10 with each row 
# so 10055 and 10075 need to be put inplace. Fill in these missing numbers and make the column an integer column
#(instead of a float column).

for i in range(0, df['FlightNumber'].count() + 1):
    if pd.isnull(df.loc[i,'FlightNumber']):
        df.loc[i, 'FlightNumber'] = df.loc[i-1, 'FlightNumber']+10
print(df)
```

```python
convert_dict = {'FlightNumber': int} 
  
df = df.astype(convert_dict) 
print(df.dtypes) 
```

```python
# 2. The From_To column would be better as two separate columns! Split each string on the underscore delimiter 
# _ to give a new temporary DataFrame with the correct values. Assign the correct column names to this 
# temporary DataFrame.

df[['From','To']]=df.From_To.str.split("_",expand=True)
#df[['First','Last']]=df.From_To.str.split("_",expand=True) 

df=df[['From_To','From','To','FlightNumber','RecentDelays','Airline']]
print(df) 
```

```python
# 3. Notice how the capitalisation of the city names is all mixed up in this temporary DataFrame. 
#Standardise the strings so that only the first letter is uppercase (e.g. "londON" should become "London".)
```

```python
df['From'] = df['From'].str.capitalize() 
df['To'] = df['To'].str.capitalize() 
df
```

```python
# 4. Delete the From_To column from df and attach the temporary DataFrame from the previous questions.
del df['From_To']
df
```

```python
#5. In the RecentDelays column, the values have been entered into the DataFrame as a list. We would like each 
#first value in its own column, each second value in its own column, and so on. If there isn't an Nth value, 
# the value should be NaN.
# Expand the Series of lists into a DataFrame named delays, rename the columns delay_1, delay_2, etc. 
# and replace the unwanted RecentDelays column in df with delays.
a=df['RecentDelays'].value_counts().index.tolist()
#print(a)
c=[]
for i in a:
    b=len(i)
    #print("\t",b)
    c.append(b)
max_columns=max(c)  
#print(max_columns)
Flight_Delays = pd.DataFrame(df['RecentDelays'].values.tolist(), columns=['Delay1','Delay2','Delay3'])  
#Variable Max Columns Returned 3. Co created 3 columns
#Flight_Delays
df1 = df.merge(Flight_Delays, left_index=True, right_index=True)
df1
```
